## Observability Overview

Observability refers to tools and practices that help developers understand software behavior and improve it. Common data collected includes:

**Metrics**: System metrics (e.g., CPU, memory, GPU memory, disk, and network per node), cluster metrics (e.g., node count and failures), and application-specific metrics.

<img src="https://docs.anyscale.com/assets/images/metrics-oa-ui-df6becb28765867764f502735fede48d.png" width="70%" loading="lazy"> 

**Profiling**: Tools to diagnose performance issues, out-of-memory errors, hangs, and other problems. Profilers analyze performance by sampling resource usage (e.g., CPU, GPU, and memory profiling).

<img src="https://anyscale-materials.s3.us-west-2.amazonaws.com/apple/tensorboard_trace_data.png" width="70%" loading="lazy"> 

**Tracing**: Mechanisms that follow execution flow to diagnose bugs, performance bottlenecks, and unexpected behavior.

<img src="https://docs.anyscale.com/assets/images/service-upstream-downstream-traces-be00d16870c6bda6507ee78f394d2c2e.png" width="70%" loading="lazy"> 

**Logs**: Records supporting monitoring and debugging (e.g., system and application logs).

<img src="https://docs.anyscale.com/assets/images/services-logs-8b9d2ef7fe8afecfa059a41cfd9bffe1.png" width="70%" loading="lazy"> 

**Events**: System-level operations handled by the operating system (OS) on behalf of applications.

<img src="https://docs.anyscale.com/assets/images/debugging_ray_oom_02-836fdee718df9ffd96a173277b7f9a91.png" width="70%" loading="lazy"> 

These data can be collected and visualized using systems such as Prometheus, OpenTelemetry, Grafana, Datadog, and so on.

Ray integrates with Prometheus, OpenTelemetry, Grafana, and other third-party tools to provide web dashboards for monitoring and debugging Ray applications. Anyscale enhances this telemetry by persisting and contextualizing the data. The next section covers local setup so examples in subsequent courses can be run and viewed locally.
